```python
import matplotlib.pyplot as plt
```

# Pie chart

```python
areas=['Marketing','Sales','Development','HR','Customer support']
budget=[6,7,10,5,4]
plt.pie(budget,labels=areas)
plt.show()
```

```python
x=[10,67,74,24,81,53]
y=[45,27,35,15,90,74]
plt.scatter(x,y)
```

```python
product_id=[1,2,3,4,5,6,7,8]
product_price_online=[20,25,65,37,40,78,95,66]
product_price_offline=[50,55,70,47,50,80,100,70]
product_price_chorbazaar=[10,15,50,20,15,60,70,60]
```

```python
plt.plot(product_id,product_price_online,color='Green',label="Online Prices")
plt.plot(product_id,product_price_offline,color='Yellow',label="Offline Prices")
plt.plot(product_id,product_price_chorbazaar,color='Purple',label="Chorbazaar Prices")
plt.legend()
plt.xlabel("Product")
plt.ylabel("Price")
plt.title("Product price comparison")
plt.show()
```

```python
year=[2012,2013,2014,2015,2016,2017,2018]
price=[65000,69000,75000,55000,52000,64000,81000]
plt.bar(year,price)
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("Bike Price per year")
plt.show()
```
